```python
from bs4 import BeautifulSoup
```

```python
def trim(s):
    return ' '.join(s.split())
```

```python
def clean(l):
    r = []
    for e in l:
        if e not in [
            'Recetas de cocina',
            'Wikimedia Commons',
            'Wikcionario',
        ]:
            x = e.strip().lower().replace('_', ' ')
            x = x.split(' (')[0]
            r.append(x)
    return r
```

```python
ingredients = set()

with open('data/ingredients/es/wikipedia/wikipedia_spanish_ingredients.html') as f:
    soup = BeautifulSoup(f.read(), 'html.parser')
    elems = soup.select('div#bodyContent h2 ~ ul li > a:nth-of-type(1), \
                         div#bodyContent h3 ~ ul li > a:nth-of-type(1), \
                         div#bodyContent h4 ~ ul li > a:nth-of-type(1)')
    ingrs = map(lambda x: trim(x['title']), elems)
    clean_ingrs = clean(ingrs)
    for ci in clean_ingrs:
        print(ci)
    ingredients = ingredients.union(clean_ingrs)

with open('data/ingredients/wikipedia_spanish_ingredients.txt', 'w') as f:
    f.write('\n'.join(sorted(ingredients)))
```
